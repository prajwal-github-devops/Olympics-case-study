```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('C:\\Users\\Prajwal\\Desktop\\Datasets\\olympics.csv')

# Filter data for swimmers from 2000 to 2016
swimmers = df[(df['Sport'] == 'Swimming') & (df['Year'].between(2000, 2016))]

# Calculate the probability of different ages
age_counts = swimmers['Age'].value_counts(normalize=True)
age_probabilities = age_counts.sort_index()  # Sort by age for better visualization

# Identify the age with the highest probability
most_likely_age = age_probabilities.idxmax()
highest_probability = age_probabilities.max()

# Print the most likely age and its probability
print(f"The age with the highest probability among swimmers: {most_likely_age}")
print(f"Probability of this age: {highest_probability:.4f}")

# Plot the age distribution as a line chart
plt.figure(figsize=(12, 6))
plt.plot(age_probabilities.index, age_probabilities.values, marker='o')
plt.title('Age Distribution of Swimmers (2000-2016)')
plt.xlabel('Age')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Load the dataset
df = pd.read_csv('C:\\Users\\Prajwal\\Desktop\\Datasets\\olympics.csv')

# Filter data for India and gold medals
india_gold = df[(df['region'] == 'India') & (df['Medal'] == 'Gold')]

# Count the number of gold medals won in each event
gold_medals_per_event = india_gold['Event'].value_counts()

# Plot the distribution
plt.figure(figsize=(12, 6))
sns.histplot(gold_medals_per_event, kde=True)
plt.title('Distribution of Gold Medals Won by India Across Olympic Events')
plt.xlabel('Number of Gold Medals')
plt.ylabel('Frequency')
plt.show()

# Calculate skewness
distribution_skewness = skew(gold_medals_per_event)

# Print skewness
print(f"Skewness of the distribution: {distribution_skewness:.2f}")

# Determine skew direction
if distribution_skewness > 0:
    print("The distribution is right-skewed (positively skewed).")
elif distribution_skewness < 0:
    print("The distribution is left-skewed (negatively skewed).")
else:
    print("The distribution is symmetric.")
```

```python
import numpy as np
import pandas as pd

# Load the dataset
olympics = pd.read_csv('C:\\Users\\Prajwal\\Desktop\\Datasets\\olympics.csv')

# Filter the data for Shooting competition
shooting_ages = olympics[olympics['Sport'] == 'Shooting']['Age'].dropna()

# Calculate the mean and standard deviation of the age for Shooting athletes
mean_age = shooting_ages.mean()
std_age = shooting_ages.std()

# Standardize the ages
standardized_ages = (shooting_ages - mean_age) / std_age

# Calculate the mean and standard deviation of the standardized age data
standardized_mean = standardized_ages.mean()
standardized_std = standardized_ages.std()

# Print the results
print(f"Standardized Mean: {standardized_mean:.2f}")
print(f"Standardized Standard Deviation: {standardized_std:.2f}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the dataset
olympics = pd.read_csv('C:\\Users\\Prajwal\\Desktop\\Datasets\\olympics.csv')

# Calculate the number of male and female athletes
total_athletes = len(olympics)
male_athletes = len(olympics[olympics['Sex'] == 'M'])
female_athletes = len(olympics[olympics['Sex'] == 'F'])

# Print counts
print(f"Number of male athletes: {male_athletes}")
print(f"Number of female athletes: {female_athletes}")

# Build hypothesis
# Null Hypothesis (H0): The proportion of male athletes is 85%
# Alternative Hypothesis (H1): The proportion of male athletes is not 85%

# Expected proportions
expected_male_proportion = 0.85
expected_female_proportion = 0.15

# Expected counts based on the null hypothesis
expected_male_count = expected_male_proportion * total_athletes
expected_female_count = expected_female_proportion * total_athletes

# Observed counts
observed_counts = [male_athletes, female_athletes]
expected_counts = [expected_male_count, expected_female_count]

# Perform Chi-Square Test
chi2_stat, p_value = chi2_contingency([observed_counts, expected_counts])[:2]

# Print the results
print(f"Chi-Square Statistic: {chi2_stat:.2f}")
print(f"P-Value: {p_value:.2f}")

# Determine if the study's claim is correct
alpha = 0.05  # Significance level
if p_value < alpha:
    print("There is a significant difference in the proportion of male athletes. The study's claim is not supported.")
else:
    print("There is no significant difference in the proportion of male athletes. The study's claim is supported.")
```
